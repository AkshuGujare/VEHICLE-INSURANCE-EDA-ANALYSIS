```python
sns.set(style="whitegrid",palette="pastel")
plt.rcParams['figure.figsize']=(10,6)
sns.histplot(df['Age'],kde=True,bins=30,color="green")
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.histplot(df['Annual_Premium'],kde=True,bins=30,color="pink")
plt.title("Annual Premium Distribution")
plt.show()
```

```python
response_counts = df['Response'].value_counts()
plt.pie(response_counts,labels=response_counts.index,
autopct="%1.1f%%",colors=["#6C88C4","#F2D4CC"],startangle=90,wedgeprops=dict(width=0.4))
plt.title('Distribution of Response')
plt.show()
```

```python
sns.countplot(data=df,x="Gender",hue="Response",palette=["#FFA477","#D5EDE7"])
plt.title("Gender's Claim")
```

```python
cross_tab = pd.crosstab(df['Vehicle_Age'], df['Response'], normalize="index")*100
plt.figure(figsize=(7,4))
cross_tab.plot(kind="bar", stacked=True, color=["#66b3ff","#99ff99"], figsize=(10,4), edgecolor="black")
plt.title("Claims by Vehicle Age (%)", fontsize=10, fontweight="bold")
plt.xlabel("Vehicle Age")
plt.ylabel("Percentage of Customers")
plt.legend(title="Response", labels=["No (0)", "Yes (1)"])
plt.xticks(rotation=0)
plt.show()
```

```python
region_counts = df['Region_Code'].value_counts()
plt.bar(region_counts.index,region_counts.values,
        color=sns.color_palette("viridis",15))
plt.xlabel("Region Code")
plt.ylabel("Count")
plt.title("Distribution of Region Codes")
plt.show()
```

```python
plt.figure(figsize=(10,6))
df['Policy_Sales_Channel'].value_counts().head(15).plot.barh(color=sns.color_palette("viridis", 15))
plt.title("Top 15 Customer Regions", fontsize=14, fontweight="bold")
plt.xlabel("Number of Customers")
plt.ylabel("Region Code")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(7,4))
cross_tab = pd.crosstab(df['Vehicle_Damage'], df['Response'], normalize="index")*100
cross_tab.plot(kind="bar", stacked=True, color=["#303030","#737373"], figsize=(10,4), edgecolor="black")
plt.title("Claims by Vehicle Damage (%)", fontsize=10, fontweight="bold")
plt.xlabel("Vehicle Damage")
plt.ylabel("Percentage of Customers")
plt.legend(title="Response", labels=["No (0)", "Yes (1)"])
plt.xticks(rotation=0)
plt.show()
```
